## 4.4 - Representing CDFs

```python
import nsfg
import thinkstats2
import thinkplot
import numpy as np
```

```python
preg = nsfg.ReadFemPreg()
live = preg[preg.prgoutcome == 1]
firsts = live[live.birthord == 1]
others = live[live.birthord != 1]
```

```python
cdf = thinkstats2.Cdf(live.prglngth, label='prglngth')
```

```python
thinkplot.Cdf(cdf)
thinkplot.Show(xlabel='weeks', ylabel='CDF', axis=[0, 50, 0 , 1])
```

## 4.5 - Compating CDFs

```python
first_cdf = thinkstats2.Cdf(firsts.totalwgt_lb, label='firsts')
other_cdf = thinkstats2.Cdf(others.totalwgt_lb, label='others')
```

```python
thinkplot.PrePlot(2)
thinkplot.Cdfs([first_cdf, other_cdf])
thinkplot.Show(xlabel='weight (lbs)', ylabel='CDF', axis=[0, 16, 0, 1])
```

## 4.7 - Random numbers

```python
weights = live.totalwgt_lb

cdf = thinkstats2.Cdf(weights, label='totalwgt_lb')
```

```python
sample = np.random.choice(weights, 100, replace=True)
ranks = [cdf.PercentileRank(x) for x in sample]
```

```python
rank_cdf = thinkstats2.Cdf(ranks)

thinkplot.Cdf(rank_cdf)
thinkplot.Show(xlabel='percentile rank', ylabel='CDF', axis=[0, 100, 0, 1])
```
